```python
import pandas as pd
import requests,json
import io
```

```python
train = pd.read_csv('datasets/train.csv')
test = pd.read_csv('datasets/test.csv')
```

```python
train_=train[['make','model','trim','body','transmission']]
vin_list = train[train_.isna().any(axis=1)]['vin']
```

```python
%%time
index = range(0,1)
df = pd.DataFrame(index=index, columns=df.columns)
# df=df.drop(axis=None)
df.drop(index=0)
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/DecodeVINValuesBatch/'
step=50

for i in range(1,len(vin_list),step):
    post_fields = {'format': 'json', 'data':';'.join(vin_list[i-1:i+step-1])}
    r = requests.post(url, data=post_fields)
    
    x = r.json()
    df_temp = pd.DataFrame(x['Results'])
    df=pd.concat([df,df_temp],axis=0)
#     df.columns
    print (i)
```

```python
df.drop(index=0).to_csv('train_miss.csv', sep='\t')
```
